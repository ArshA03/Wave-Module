# Analysing Peaks In Healthy State For Bearing 3

These Data Are Collected By two Channels. We'll analyse each channel respectively.
The accelerometers related to these channels are perpendicular to each other so these channels have a Phase Difference of 90 degrees.

### First Channel

Input Data

```python
Hdata1 = Input(r'D:\Academics\Thesis\DataBases\IMS\1st_test\2003.10.22.12.06.24')
Hdata1 = Hdata1['channel5'].to_numpy()  # Channel 5 is for Bearing 3
```

FFT Spectrum

```python
# Bearing 3 vibration data is in channel 5 and 6
Hwave1 = wave(data=Hdata1, Fs=20000)  # this variable is in wave class and can use it's functions
Hfreq1, Hmag1= Hwave1.FFT()
```

Plotting

```python
fig, (ax1, ax2) = plt.subplots(ncols=1, nrows=2, constrained_layout=True, figsize=(10, 6))

############## Time WaveForm ##################
ax1.set_title('Time WaveForm')
ax1.plot(Hwave1.time(), Hdata1, color='k')
ax1.set_xlim(0,1)
ax1.set_ylabel('Amplitude')
ax1.set_xlabel('Time (s)')

############# FFT Spectrum ####################
ax2.set_title('FFT Spectrum')
ax2.plot(Hfreq1, Hmag1)
ax2.set_ylim(0, 0.1)
ax2.set_xlim(0, 10000)
ax2.set_xlabel('Frequency in Hertz [Hz]')
ax2.set_ylabel('Frequency Domain (Spectrum) Magnitude')
```
